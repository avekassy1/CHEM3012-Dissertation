```python
# L) literature review, M) measured, V) VP/AS = vapor pressure/aqueous solubility, R) recalculation, 
# T) thermodynamical calculation, X) original paper not available, C) citation, Q) QSPR, E) estimate, ?) unknown, W) wrong

# eliminate Wrong, Estimate, X=paper unknown, ?=unknown, Q, Thermod calc, and Recalc are not measured
# remaining is VP/AS, Measured, Literature, Citation
converted = converted[converted['htype'].isin(['V', 'M', 'L', 'C'])]
```

```python
# similarly to aqsoldb, calculate avg/geom mean/median for KH value and std deviation to see which one's best for ML
# cannot have more than one row for one species for modelling purposes innit
avg_kh = converted.groupby('species_id').mean()['Kh'].rename('avg_kh').reset_index(drop=True)
avg_kh
```

```python
converted.groupby('species_id').count()
```

```python
occurrences = converted.groupby('species_id').count()['id'].values
print(occurrences.shape)
occurrences = pd.Series(occurrences).rename('occurrences')
occurrences
```

```python
spec_ids = pd.Series(converted.groupby('species_id').count()['id'].index, dtype=int)#.rename()
spec_ids
```

```python
med_kh = converted.groupby('species_id').median()['Kh'].rename('med_kh').reset_index(drop=True)
```

```python
std_kh = converted.groupby('species_id').std()['Kh'].rename('std_kh').reset_index(drop=True)
```
