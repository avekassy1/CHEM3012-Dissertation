```python
df = pubchem.merge(chembl, how='outer', left_on='inchikey', right_on='chembl_inchikeys')
df
```

```python
df.inchikey = df.inchikey.combine_first(df.chembl_inchikeys)
df['smiles'] = df.smiles_y.combine_first(df.smiles_x) # CheMBL smiles are priorities as they have some standardisation protocols mentioned on their website
# https://chembl.gitbook.io/chembl-interface-documentation/frequently-asked-questions/chembl-data-questions
```
