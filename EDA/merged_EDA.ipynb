# ML

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.metrics import mean_squared_error, r2_score
# Basic regressors
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
# Multitasking
from sklearn.multioutput import MultiOutputRegressor
from sklearn.neural_network import MLPRegressor

```

```python
# Combined output
Y = pd.concat([df_logS, df_logHcp], axis=1)
Y.head()
```

```python
# Split
train_X, test_X, train_Y, test_Y = train_test_split(descriptors, Y, test_size=0.3) #random_state=42
```

```python
print(train_X.shape, test_X.shape, train_Y.shape, test_Y.shape)
```

```python
train_y_logs = train_Y.LogS
train_y_logkh = train_Y.Kh_first
test_y_logs = test_Y.LogS
test_y_logkh = test_Y.Kh_first
```

```python
feature_names = list(train_X.columns)
```

```python
# SINGLETASKING
sc = StandardScaler()
train_X = sc.fit_transform(train_X)
test_X = sc.transform(test_X)
```

```python
# Sol
regressors = [LinearRegression(), SVR(), DecisionTreeRegressor(), RandomForestRegressor(), GradientBoostingRegressor()]
models = ['Linear Regression', 'Support Vector Regressor', 'Decision Tree Regressor', 'Random Forest Regressor',
            'Gradient Boosting Regressor']
train_r2s = []
test_r2s = []
rmses = []

for reg in regressors:

    reg.fit(train_X, train_y_logs) # Expand with k-fold cross val or gridsearchCV in future

    pred_train = reg.predict(train_X)
    pred_test = reg.predict(test_X)

    train_r2s.append(r2_score(train_y_logs, pred_train))
    test_r2s.append(r2_score(test_y_logs, pred_test))
    rmses.append(mean_squared_error(test_y_logs, pred_test, squared=False))

    # Plot    

scores_sol = pd.DataFrame({'Model': models, 'Train r2': train_r2s, 'Test r2': test_r2s, 'RMSE': rmses})
scores_sol  
```

```python
# MULTITASKING

# Scale
sc = StandardScaler()
train_X = sc.fit_transform(train_X)
test_X = sc.transform(test_X)
```

```python
# Fit
regressors = [LinearRegression(), SVR(), DecisionTreeRegressor(), RandomForestRegressor(), GradientBoostingRegressor()]
models = ['Linear Regression', 'Support Vector Regressor', 'Decision Tree Regressor', 'Random Forest Regressor',
            'Gradient Boosting Regressor']
train_r2s = []
test_r2s = []
rmses = []
```

```python
for reg in regressors:

    mulreg = MultiOutputRegressor(reg)
    mulreg.fit(train_X, train_Y) # Expand with k-fold cross val or gridsearchCV in future

    pred_train = mulreg.predict(train_X)
    pred_test = mulreg.predict(test_X)

    train_r2s.append(r2_score(train_Y, pred_train))
    test_r2s.append(r2_score(test_Y, pred_test))
    rmses.append(mean_squared_error(test_Y, pred_test, squared=False))

    # Plot    

scores = pd.DataFrame({'Model': models, 'Train r2': train_r2s, 'Test r2': test_r2s, 'RMSE': rmses})
```
